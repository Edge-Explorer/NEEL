# Notebook 06: Analytics Engine

Purpose:
- Generate weekly and monthly analytics from activity data
- Create human-readable insights
- Prepare summaries for LLM reasoning and supervisor checks

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
time_df = pd.read_csv("../data/processed/time_features.csv")
study_df = pd.read_csv("../data/processed/study_features.csv")
habit_df = pd.read_csv("../data/processed/habit_features.csv")
```

```python
avg_time = time_df[
    ["work_minutes", "leisure_minutes", "exercise_minutes", "sleep_minutes"]
].mean()

plt.figure()
avg_time.plot(kind="pie", autopct="%1.1f%%")
plt.title("Average Weekly Time Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure()
time_df["productivity_score"].plot()
plt.title("Productivity Trend")
plt.xlabel("Days")
plt.ylabel("Productivity Score")
plt.show()
```

```python
plt.figure()
plt.scatter(
    study_df["study_effort_score"],
    study_df["exam_score"]
)
plt.xlabel("Study Effort Score")
plt.ylabel("Exam Score")
plt.title("Effort vs Outcome")
plt.show()
```

```python
habit_df["habit_balance"] = (
    habit_df["recovery_score"] - habit_df["distraction_load"]
)

habit_df["habit_balance"].describe()
```

```python
summary = {
    "avg_work_minutes": round(avg_time["work_minutes"], 1),
    "avg_leisure_minutes": round(avg_time["leisure_minutes"], 1),
    "avg_sleep_minutes": round(avg_time["sleep_minutes"], 1),
    "avg_exercise_minutes": round(avg_time["exercise_minutes"], 1),
    "mean_productivity": round(time_df["productivity_score"].mean(), 2),
}
summary
```
